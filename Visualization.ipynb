```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
```

```python
mydata = pd.read_csv('data.csv')
```

```python
mydata.describe()
```

```python
# dataframe column year, area_id, variable_ID is treated as integer 
# to make it a category
mydata.year = mydata.year.astype('category')
mydata.area_id = mydata.area_id.astype('category')
mydata.variable_id = mydata.variable_id.astype('category')
```

```python
list1 =[]
list2=list()

for k in mydata.year.cat.categories:
    list1.append(mydata[mydata.year==k].area_id)
    list2.append(k)
plt.hist(list1,bins=20,stacked=True,label=list2)
plt.rcParams['figure.figsize']=10,12
plt.legend()
plt.show()
```
