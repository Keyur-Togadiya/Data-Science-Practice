```python
import vlc
from tkinter import *
import datetime

flag=0
press=0
class Mains:

    def __init__(self,master):
        self.master = master
        self.frame = Frame(self.master, width=600, height=1000, bg='cornflowerblue')
        self.frame.pack()

        self.txt = Text(self.frame, width=77, height=5, wrap=WORD, bg='lightgrey')
        self.txt.place(x=25,y=25)
        
        self.b1 = Button(self.frame,text='1', padx=20, pady=20, bg='white', command = lambda:[self.text_change()])
        self.b1.place(x=125, y=150)
        self.b2 = Button(self.frame,text='2', padx=20, pady=20, bg='white',command = lambda:[self.date1()])
        self.b2.place(x=275, y=150)
        self.b3 = Button(self.frame,text='3', padx=20, pady=20, bg='white')
        self.b3.place(x=425, y=150)
        self.b4 = Button(self.frame,text='4', padx=20, pady=20, bg='white')
        self.b4.place(x=125, y=275)
        self.b5 = Button(self.frame,text='5', padx=20, pady=20, bg='white')
        self.b5.place(x=275, y=275)
        self.b6 = Button(self.frame,text='6', padx=20, pady=20, bg='white')
        self.b6.place(x=425, y=275)
        self.b7 = Button(self.frame,text='7', padx=20, pady=20, bg='white')
        self.b7.place(x=125, y=400)
        self.b8 = Button(self.frame,text='8', padx=20, pady=20, bg='white')
        self.b8.place(x=275, y=400)
        self.b9 = Button(self.frame,text='9', padx=20, pady=20, bg='white')
        self.b9.place(x=425, y=400)
        self.b10 = Button(self.frame,text='*', padx=20, pady=20, bg='white')
        self.b10.place(x=125, y=525)
        self.b11 = Button(self.frame,text='0', padx=20, pady=20, bg='white')
        self.b11.place(x=275, y=525)
        self.b12 = Button(self.frame,text='#', padx=20, pady=20, bg='white')
        self.b12.place(x=425, y=525)

        self.banswer = Button(self.frame,bg='lightgreen', command = self.play_sound)
        self.banswer.place(x=180, y=600)
        self.file1 = PhotoImage(file="answer.png")
        self.banswer.config(image=self.file1, compound=CENTER)
        self.file2 = self.file1.subsample(7, 7)
        self.banswer.config(image=self.file2)

        self.breject = Button(self.frame,bg='red', command = self.stop_sound)
        self.breject.place(x=330, y=600)
        self.file3 = PhotoImage(file="reject.png")
        self.breject.config(image=self.file3, compound=CENTER)
        self.file4 = self.file3.subsample(7, 7)
        self.breject.config(image=self.file4)
        

    def text_change(self):
        global text
        global flag
        if(flag==0):
            self.txt.delete('1.0',END)
            self.show_data1()
            flag+=1
        elif(flag==1):
            self.show_data2()
            flag+=1
        elif(flag==2):
            self.show_data3()
            flag+=1
        elif(flag==3):
            self.show_data4()
            flag+=1
        else:
            self.date()

    def date(self):
        self.sound_file.stop()
        self.txt.delete('1.0',END)
        self.today = datetime.date.today()
        self.tomorrow = self.today + datetime.timedelta(days = 1)
        t = "Your appointment is booked on", self.tomorrow, '\n'
        self.txt.insert(0.0,t)

    def date1(self):

        if(flag==4):
            self.sound_file.stop()
            self.txt.delete('1.0',END)
            self.today = datetime.date.today()
            self.dtomorrow = self.today + datetime.timedelta(days = 2)
            t = "Your appointment is booked on", self.dtomorrow, '\n'
            self.txt.insert(0.0,t)

 
    def show_data(self):
        self.sound_file.stop()
        t = "Your Appointment is booked for 10:00\n"
        self.txt.insert(0.0,t)

    def show_data1(self):
        self.sound_file.stop()
        self.sound_file = vlc.MediaPlayer("hello1.mp3")
        self.sound_file.play()

    def show_data2(self):
        self.sound_file.stop()
        self.sound_file = vlc.MediaPlayer("hello2.mp3")
        self.sound_file.play()

    def show_data3(self):
        self.sound_file.stop()
        self.sound_file = vlc.MediaPlayer("hello3.mp3")
        self.sound_file.play()

    def show_data4(self):
        self.sound_file.stop()
        self.sound_file = vlc.MediaPlayer("hello4.mp3")
        self.sound_file.play()
    
    def play_sound(self):
        self.txt.delete('1.0',END)
        self.sound_file = vlc.MediaPlayer("hello.mp3")
        self.sound_file.play()

    def stop_sound(self):
        self.sound_file.stop()



if __name__ == '__main__':
    root = Tk()
    root.title('IVR')
    b = Mains(root)
    root.mainloop()
    
'''self.txt.delete('1.0',END)
            t1 =  "Appointment booked"
            self.txt.insert(0.0,t1)'''



```
