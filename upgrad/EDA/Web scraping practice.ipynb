```python
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
```

```python
myurl = "https://www.flipkart.com/search?q=sennheiser&as=on&as-show=on&otracker=AS_Query_OrganicAutoSuggest_2_8_na_na_na&otracker1=AS_Query_OrganicAutoSuggest_2_8_na_na_na&as-pos=2&as-type=RECENT&suggestionId=sennheiser&requestId=c9c6fe71-62b0-43bf-bbc6-4e16bafb75ba&as-searchtext=sennhise&sort=price_desc"
getdata = uReq(myurl)
m = getdata.read()
getdata.close()
```

```python
html_parse1 = soup(m)
```

```python
contain = html_parse1.findAll("div", {'class': "_3liAhj"})
print(len(contain))
```

```python
with open("flipkart.csv","w") as f:

    headers = "Product, Price, Features \n"
    f.write(headers)
    
    for block in contain:
        product_s = block.findAll('a', {'class': '_2cLu-l'})
        product = product_s[0].text
        
        price_s = block.findAll('div', {'class': '_1vC4OE'})
        price = price_s[0].text
        price = price.replace(',','')
        price = price[1:]
        
        feature_s = block.findAll('div', {'class': '_1rcHFq'})
        feature = feature_s[0].text
        if "," in feature:
            ind = feature.index(',')
            feature = feature[:ind]

        f.write(product + ',' + price + ',' + feature  + "\n")
```
