## Python Libraries - Pandas - Time Series

```python
# Import the required libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# read the dataset 'weather_data.csv'
weather = pd.read_csv('weather_data.csv')
```

```python
# Check the created dataframe
weather.head()
```

```python
# Read the dataframe with Date_Time variable in the correct form
weather = pd.read_csv('weather_data.csv', parse_dates=['Date_Time'])

# Check the created dataframe
weather.head()
```

```python
# Read the dataframe with Date_Time variable in the correct form and as Index
weather = pd.read_csv('weather_data.csv', parse_dates=True, index_col=['Date_Time'])

# Check the created dataframe
weather
```

```python
# Delete the index label
weather.index.name
```

```python
del weather.index.name
```

```python
# Fetch the entries for the date: 'February 5, 2010'
weather.loc['2010-02-05']
```

```python
weather.loc['February 5,2010']
```

```python
# Slice all the entries between 9 AM and 10 PM on 11th October
weather.loc['2010-10-11 9:00:00' : '2010-10-11 22:00:00']
```

```python
# Slice all the entries between 15th and 31st October
weather.loc['2010-10-15': '2010-10-31']
```

```python
# Plot the trend of the data for the entries obtained above
dec = weather.loc['2010-10-15': '2010-10-31']
dec['Temperature'].plot(kind='line')
plt.show()
```
